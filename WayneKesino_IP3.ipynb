##1.CLEANING THE DATA

```
#Renaming the PRODUTC and the DATETIME columns on telcom_dataset to PRODUCT and DATE_TIME 
df2.rename(columns = {'PRODUTC': 'PRODUCT' , 'DATETIME': 'DATE_TIME'}, inplace = True)
df2
```

```
#Renaming the columns 'DW_A_NUMBER' AND 'DW_B_NUMBER' on the telcom_datset2 dataframe
#to 'DW_A_NUMBER_INT' AND 'DW_B_NUMBER_INT'
df3.rename(columns = {'DW_A_NUMBER' : 'DW_A_NUMBER_INT' , 'DW_B_NUMBER' : 'DW_B_NUMBER_INT'} , inplace=True)
df3
```

```
#Renaming the ‘CELLID’’ and the ‘SIET_ID’ columns on the telcom_dataset3 dataframe to ‘CELL_ID’ and ‘SITE_ID’ respectively.
df4.rename(columns = {'CELLID' : 'CELL_ID' , 'SIET_ID' : 'SITE_ID'} , inplace=True)
df4
```

##2.MERGING DATASETS, CONSTRUCTING NEW COLUMNS AND SPLITTING AND DELETING.

```
#Merging the datasets from the three days 
df7 = pd.concat([df2, df3, df4] , ignore_index=True)
df7
```

```
#Splitting the DATE_TIME column into two 
df7[['DATE','TIME']] = df7.DATE_TIME.str.split(expand=True)
df7
```

```
#Deleting the original DATE_TIME column 
del df7['DATE_TIME']
df7.head(60)
```

```
#Merging the combined dataset(df7) and the cell_geo dataset
df8 = df7.merge(df5, how = 'right', on = 'CELL_ID')
df8
```

##3.GETTING THE MOST USED CITIES IN THE 3 DAYS

```
#Calculating the most used cities
df5['VILLES'].value_counts().head(10)
```

##4.THE CITIES THAT WERE THE MOST USED DURING BUSINESS AND HOME HOURS

```
#Grouping the dataset to determine the number of times they were in service using the status column
df9 = df5.groupby(['VILLES'], as_index = False)["STATUS"].count()
#Sorting them in descending order in order to determine the most used cities
df9.sort_values(["STATUS","VILLES"], axis = 0, ascending=False, inplace=True)
df9.head(15)
```
